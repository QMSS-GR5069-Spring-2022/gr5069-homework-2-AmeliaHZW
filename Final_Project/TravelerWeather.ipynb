```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
%matplotlib inline
```

```python
import requests
from requests.exceptions import HTTPError
import os
import json
from dotenv import load_dotenv
```

```python
from dotenv import load_dotenv
load_dotenv('API_env')
key = os.getenv('API_key')
```

```python
city = ['New York','London', 'Shanghai', 'Washington D.C.']
```

```python
city_name = 'Shanghai'
```

```python
r = requests.get('http://api.openweathermap.org/data/2.5/forecast?q='+city_name+'&APPID='+key)
```

```python
OWM_forecast_json = r.json()
print(type(OWM_forecast_json))
print(OWM_forecast_json.keys())
```

```python
OWM_forecast_json['list']
```

```python
#Extract columns and transform into dataframe
df_m = pd.DataFrame()
i = 0
for i in range(len(OWM_forecast_json['list'])):
    df_m = pd.concat([df_m, pd.DataFrame(OWM_forecast_json['list'][i]['main'], index = [i])])
    i = i+1
    

df_we = pd.DataFrame()
i = 0
for i in range(len(OWM_forecast_json['list'])):
    df_we = pd.concat([df_we, pd.DataFrame(OWM_forecast_json['list'][i]['weather'], index = [i])])
    i = i+1


df_wi = pd.DataFrame()
i = 0
for i in range(len(OWM_forecast_json['list'])):
    df_wi = pd.concat([df_wi, pd.DataFrame(OWM_forecast_json['list'][i]['wind'], index = [i])])
    i = i+1

df_dt = pd.DataFrame()
i = 0
for i in range(len(OWM_forecast_json['list'])):
    df_dt = pd.concat([df_dt, pd.Series(OWM_forecast_json['list'][i]['dt_txt'], index = [i])])
    i = i+1    
   
        
df = pd.concat([df_m,df_we,df_wi,df_dt], axis = 1)
    

```

```python
df = df.rename(columns={0:'local_time'})
```

```python
df['temp_C'] = df['temp'] - 273.15
```

```python
df['temp_F'] = 1.8*(df['temp']-273)+32
```

```python
fig,ax = plt.subplots(1,1,figsize=(24,10))
```
